```python
import pandas as pd 

diabetes_messy_df = pd.read_csv("diabetes_messy.csv")
diabetes_messy_df.head()
```

```python
#Making lookup table

#extract only necessary parts
table1 = diabetes_messy_df.iloc[2:4].iloc[:,9:16]

#transpose
table1 = table1.T

#change column name for convenience
table1.columns = ['Country', 'Study_started']
table1 = table1.iloc[1:]
```

```python
table1
```

```python
repl = table1.set_index('Country')['Study_started'].to_dict()
```

```python
diabetes_messy_df = diabetes_messy_df.rename({'United States': 'Study_started'}, axis=1) 
```
